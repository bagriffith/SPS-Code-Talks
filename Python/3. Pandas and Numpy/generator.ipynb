```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
stars = pd.DataFrame()
```

```python
# Background
frame = pd.DataFrame()

N = 100

frame['ra'] = 28.2 + np.random.uniform(0, .1, N)
frame['dec'] = 56.9 + np.random.uniform(0, .1, N)
frame['T'] = np.random.gamma(3, 1000, N) + 2500

stars = stars.append(frame)
```

```python
# Cluster 1
frame = pd.DataFrame()

N = 100

frame['ra'] = 28.2 + np.random.normal(.04, .01, N)
frame['dec'] = 56.9 + np.random.normal(.07, .01, N)
frame['T'] = np.random.normal(5500, 250, N)

stars = stars.append(frame)
```

```python
plt.style.use('dark_background')

plt.scatter(stars['ra'],
            stars['dec'],
            c=stars['T'],
            cmap=plt.get_cmap('RdBu'),
            norm=mpl.colors.Normalize(vmin=1000, vmax=12000))
```
